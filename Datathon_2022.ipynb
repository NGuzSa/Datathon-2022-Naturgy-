```
import pandas as pd
train= pd.read_csv ("/content/train.csv") 
test= pd.read_csv ("/content/test.csv") 
train.head()
train.info() 
```

El código anterior nos da información de las variables con valores nulos en el data set de train.

Valores nulos se conocen como missing data. Buscar método para completar missing data (i.e. basados es estadística clásica)

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```
train["State_Factor"].value_counts()
```

```
train["facility_type"].value_counts()
```

```
from sklearn.preprocessing import LabelEncoder 
Le= LabelEncoder ()
```

```
for i in train.columns:
  if train[i].dtypes=="object":
    train[i]=Le.fit_transform(train[i])
train.head()
```

```
for i in test.columns:
  if test[i].dtypes=="object":
    test[i]=Le.fit_transform(test[i])
test.head()
```

El código anterior es para hacer un bucle que recorra todas las columnas para encontrar una variable categorica y que la transforme en variable numérica

```
from sklearn.impute import SimpleImputer
```

```
train_year_built_imputation_value = 2022
train_energy_star_rating_imputation_value = train['energy_star_rating'].mean()
train_direction_max_wind_speed_imputation_value = train['direction_max_wind_speed'].mean()
train_direction_peak_wind_speed_imputation_value = train['direction_peak_wind_speed'].mean()
train_max_wind_speed_imputation_value = train['max_wind_speed'].mean()
train_days_with_fog_speed_imputation_value = train['days_with_fog'].mean()

train['year_built'] =train['year_built'].replace(np.nan, train_year_built_imputation_value)
train['energy_star_rating']=train['energy_star_rating'].replace(np.nan, train_energy_star_rating_imputation_value)
train['direction_max_wind_speed']= train['direction_max_wind_speed'].replace(np.nan, train_direction_max_wind_speed_imputation_value)
train['direction_peak_wind_speed']= train['direction_peak_wind_speed'].replace(np.nan, train_direction_peak_wind_speed_imputation_value)
train['max_wind_speed']=train['max_wind_speed'].replace(np.nan, train_max_wind_speed_imputation_value)
train['days_with_fog']=train['days_with_fog'].replace(np.nan, train_days_with_fog_speed_imputation_value)
```

```
test['year_built'] =test['year_built'].replace(np.nan, train_year_built_imputation_value)
test['energy_star_rating']=test['energy_star_rating'].replace(np.nan, train_energy_star_rating_imputation_value)
test['direction_max_wind_speed']= test['direction_max_wind_speed'].replace(np.nan, train_direction_max_wind_speed_imputation_value)
test['direction_peak_wind_speed']= test['direction_peak_wind_speed'].replace(np.nan, train_direction_peak_wind_speed_imputation_value)
test['max_wind_speed']=test['max_wind_speed'].replace(np.nan, train_max_wind_speed_imputation_value)
test['days_with_fog']=test['days_with_fog'].replace(np.nan, train_days_with_fog_speed_imputation_value)
```

El código anterior cambia los missing data por los valores que queremos en función de cada caso. Hemos determinado que year_built se cambien por 2022 y en el resto de los casos por la media, tanto para train como test.

```
train.describe().style.background_gradient()
```

Ahora vemos los nuevos valores de las variables que tenían missing values.

```
import copy 
from sklearn.preprocessing import StandardScaler
y = train["site_eui"]
train = train.drop(["site_eui","id"],axis =1)
test = test.drop(["id"],axis =1)
trainnames = copy.deepcopy(train)
scaler = StandardScaler()
train = scaler.fit_transform(train)
test = scaler.fit_transform(test)
```

Lo que hemos hecho ahora es aplicar lo que denominan "Feature scaling" para normalizar los datos que queremos dentro de un rango en particular.

```
train
```

Aplicamos el modelo XGboost ya que es uno de los modelos más empleados para tratar datos estructurados y de manera rápida. 

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, y, test_size = 0.2, random_state = 50)
```

```
import xgboost
```

```
xgboost_model = xgboost.XGBRegressor(n_estimators=100, learning_rate=0.03, gamma=0, subsample=0.75,
                           colsample_bytree=0.4, max_depth=3)
xgboost_model.fit(X_train,y_train)
```

```
from sklearn.metrics import explained_variance_score
predictions = xgboost_model.predict(X_test)
print(explained_variance_score(predictions,y_test))
```

```
res = xgboost_model.predict(test)
sub = pd.read_csv("/content/sample_solution.csv")
sub["site_eui"] = res
sub.to_csv("submission.csv", index = False)
```
